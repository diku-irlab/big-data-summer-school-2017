The first example shows how to generate data given Equation 1.1 from the book on p. 9,  
  
$$y_{t} = \alpha + \beta y_{t-1} + \epsilon_{t}$$
  
where,  
  
$$y_{0}=1, \beta=0.95, \alpha=-0.25$$
  
and $\epsilon_{t}$ is a random number from a uniform distribution in the range $[-1,1]$.  

```python
# From pages 9-10
import pandas as pd
import numpy as np
seed = 2016
np.random.seed(seed)
y_0 = 1
alpha = -0.25
beta = 0.95
y = pd.Series(y_0)
num = 10
for i in range(num):
    y_t = alpha + (beta * y_0) + np.random.uniform(-1,1)
    y = y.set_value(i, y_0)
    y_0 = y_t

print "Values of y are:"
for val in y.values:
    print "  {:0.04f}".format(val)
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')

# Figure 1.2 (p.11)
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)
y.plot(xlim=(0, 9), ylim=(0.6, 2.2))
plt.title("Figure 1.2: Simulated time series")
plt.xlabel("Time step")
plt.ylabel("Value of $y$")
plt.show()
```

```python
np.random.seed(seed)
y_0 = 1
y = pd.Series(y_0)
num = 1000
for i in range(num):
    y_t = alpha + (beta * y_0) + np.random.uniform(-1,1)
    y = y.set_value(i, y_0)
    y_0 = y_t

# Figure 1.2 (p. 11)
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)
y.plot(xlim=(0, 1000), ylim=(-12, 4))
plt.title("Figure 1.3: 1000 simulated time series observations")
plt.xlabel("Time step")
plt.ylabel("Value of $y$")
plt.show()
```
